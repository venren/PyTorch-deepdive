```python
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
```

```python
# Download MNIST dataset
train_dataset = datasets.MNIST(root='./data', 
                             train=True, 
                             transform=transforms.ToTensor(),
                             download=True)

test_dataset = datasets.MNIST(root='./data', 
                            train=False, 
                            transform=transforms.ToTensor(),
                            download=True)
```

```python
# Get the number of datapoints in the training dataset
print(f"Number of datapoints in train_dataset: {len(train_dataset)}")

# Get the shape of the first data sample in the training dataset
sample_data, sample_label = train_dataset[0]
print(f"Shape of the first data sample: {sample_data.shape}")
```

```python
for image, label in train_dataset:
    print(label)
    print(image.shape)
    break
```

```python
### Iterate and Display Images ###
counter = 0
for image, label in train_dataset:
    if counter == 6:
        break
    image = image.squeeze().numpy()
    print("Image Label:", label)
    plt.imshow(image, cmap="gray")
    plt.show()
    counter += 1
```
